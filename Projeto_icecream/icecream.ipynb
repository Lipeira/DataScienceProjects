```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IceCreamData.csv')
```

```python
df
```

```python
df = df.head(12)
df = df.sort_values('Temperature')
df
```

```python
plt.title('Ice Cream Truck')

plt.plot(df['Temperature'], df['Revenue'])

plt.show()
```

```python
plt.title('Ice Cream Truck')

plt.plot(df['Temperature'], df['Revenue'], label='US$')

plt.ylabel('Sales (US$)')
plt.xlabel('Temperatura (Fahrenheit)')

plt.legend()
plt.show()
```

```python
plt.title('Ice Cream Truck')

plt.plot(df['Temperature'], df['Revenue'], 'r.--', label='US$')

plt.ylabel('Sales (US$)')
plt.xlabel('Temperatura (Fahrenheit)')

plt.savefig('icecream_sales.pdf', format='pdf')

plt.legend()
plt.show()
```
